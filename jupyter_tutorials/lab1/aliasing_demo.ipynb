```python
%matplotlib notebook
import numpy as np, pylab as plt
```

```python
from ipywidgets import interact
fq = 1. # Hz
def pltsin(dt):
    ts = np.arange(0., 20, dt)
    plt.clf()
    plt.plot(ts, np.sin(2*np.pi*fq*ts), '.-')
    plt.xlim(0,20)
    plt.ylim(-1,1)
    plt.show()
i = interact(pltsin, dt=(.005,3,.01))
plt.figure()
pltsin(.01)
```
