```python
%matplotlib notebook
import numpy as np, pylab as plt
```

```python
from ipywidgets import interact
fq = 1. # Hz
def pltsin(dt):
    ts = np.arange(0., 20, dt)
    plt.clf()
    plt.plot(ts, np.sin(2*np.pi*fq*ts), '.-')
    plt.xlim(0,20)
    plt.ylim(-1,1)
    plt.show()
i = interact(pltsin, dt=(.005,3,.01))
plt.figure()
pltsin(.01)
```

```python
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
fig, ax = plt.subplots()

ax.set_xlim((-2, 2))
ax.set_ylim((-2, 2))
ax.grid()

line, = ax.plot([], [],'k.')

def init():
    line.set_data([], [])
    return (line,)

def animate(i):
    fq = 1.
    t = np.array([i/40.])
    x = np.cos(2*np.pi*fq*t)
    y = np.sin(2*np.pi*fq*t)
    line.set_data(x, y)
    return (line,)

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=100, interval=20, blit=True)
```
